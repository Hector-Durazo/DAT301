The matlab method plt.subplots (note 's' at the end) creates a tuple, whose first component is a new figure and the 2nd one is a numpy array containing the created subplot objects. This is convenient, as it creates nice indexing for easier access of each subplot. If the created numpy array of subplots is called axes, then, for example, axes[0,1] means subplot in the 1st row and 2nd column.

```python
fig, axes = plt.subplots(2, 3)
axes
```

#### Adjusting the spacing around subplots

```python
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=None, hspace=None)
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
#sharex=True means that all subplots have the same x-axis, which is convenient when we compare data on the same scale.
#Similarly for sharey=True
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        #accessing (i,j)-subplot (convenient indexing)
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0.05, hspace=0.05) #controlling spacing between subplots - width-space and height-space
```

### Annotations and Drawing on a Subplot

ax.text(x, y, 'Hello world!',

```python
from datetime import datetime
import pandas as pd

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
print(data)
spx = data['SP500']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak before market crash'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')

# Zoom in on 2007-2010
ax.set_xlim(['1998-01-01', '2018-01-01'])
ax.set_ylim([600, np.max(spx)])

ax.set_title('S&P500 Index')
```

##### With data that requires aggregation or summaries before making a plot, using seaborn library can simplify coding.

```python
import seaborn as sns
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```python
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
```

## `seaborn.pairplot()`

<br>

By default, this function will create a grid of Axes such that each numeric variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```

### Facet Grids and Categorical Data

```python
## use catplot() for categorical plot; used to be called factorplot()
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```

```python
plt.close('all')
sns.catplot(x='day', y='tip_pct', row='time',
               col='smoker',
               kind='bar', data=tips[tips.tip_pct < 1])
```
