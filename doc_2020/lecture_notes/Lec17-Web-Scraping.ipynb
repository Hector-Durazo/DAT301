```python
# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Set the url
url = "https://phoenix.craigslist.org/"
```

<img src="craigslist.png" width="700" height="400">

```python
# Get Url
page = requests.get(url)
```

```python
# View Content
page.content
type(page)
```

```python
# Turn response into a soup object to parse html content
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Use find method
soup.find('title')
```

```python
# Use find_all method to find all "a" tags
a = soup.find_all('a')
```

```python
a
##type(a)
```

```python
# Print first 4 a-tags
a[:4]
```

```python
# Use find method with specified attributes
#soup.find('div')
soup.find('div', attrs={'class': 'jobs'})
```

```python
# Use find method with specified attributes
nearby_cl = soup.find('ul', attrs={'class': 'acitem'})
nearby_cl
```

```python
myli = soup.find('ul', attrs={'id': 'jjj0'})
myli.find_all('a')
```

```python
nearby_list = nearby_cl.find_all('a')
```

```python
nearby_list
```

```python
# Create a list of the nearby city names
names_list = []
for i in range(len(nearby_list)):
    names_list.append(nearby_list[i].get_text())
```

```python
names_list
```

```python
# Create a list of the nearby citities links
link_list = []
for i in range(len(nearby_list)):
    link_list.append(nearby_list[i].get('href'))
```

```python
link_list
```

```python
# Create pandas dataframe with specified column name
columnHeaders = ['Nearby', 'Links']
df = pd.DataFrame(columns = columnHeaders)
df
```

```python
df = pd.DataFrame({'Nearby':pd.Series(names_list), 'Links':pd.Series(link_list)})
df
```
